```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("rohit.jpg")
plt.imshow(img)
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img_g = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_g,cmap='gray')
```

```python
img_inv = cv2.bitwise_not(img_g)
```

```python
plt.imshow(img_inv,cmap='gray')
```

```python
img_sm = cv2.GaussianBlur(img_inv, (25,25), sigmaX=0, sigmaY = 0)
plt.imshow(img_sm, cmap='gray')
```

```python
img_pencil = cv2.divide(img_g, 255-img_sm, scale=255)
```

```python
plt.imshow(img_pencil, cmap='gray')
```

```python
face_cas = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def detect_face(image):
    face_img = image.copy()
    face_cont = face_cas.detectMultiScale(face_img)

    for(x,y,w,h) in face_cont:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 4)

    return face_img
```

```python
res = detect_face(img)
plt.imshow(res, cmap='gray')
```
